```python
import pandas as pd
```

```python
brands = pd.read_csv("../data/brands.csv", sep=";")
models = pd.read_csv("../data/model_specs.csv", sep=";")
expansion = pd.read_csv("../data/expansion_markets.csv", sep=";")
```

```python
brands.head()
```

```python
df = models.merge(
    brands,
    on="brand",
    how="left"
)

df = df.merge(
    expansion,
    on="brand",
    how="left"
)

df.head()
```

```python
df.columns
```

```python
df
```

```python
df.describe()
```

```python
df.groupby("brand")["range_km"].mean()
```

```python
df.sort_values("range_km", ascending=False)
```

```python
import matplotlib.pyplot as plt

avg_range = df.groupby("brand")["range_km"].mean()

avg_range.plot(kind="bar")

plt.ylabel("Range (km)")
plt.title("Average EV Range by Brand")

plt.show()
```

```python
df.sort_values("expansion_score", ascending=False)
```

```python
df[["brand","model","price_eur","range_km","expansion_score"]]
```

```python
df["price_per_km"] = df["price_eur"] / df["range_km"]

df[["brand","model","price_per_km"]].sort_values(
    "price_per_km"
)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(
    df["price_eur"],
    df["range_km"]
)

for i, row in df.iterrows():
    plt.text(
        row["price_eur"],
        row["range_km"],
        row["brand"]
    )

plt.xlabel("Price (€)")
plt.ylabel("Range (km)")
plt.title("Price vs Range")

plt.show()
```

```python
df["price_eur"].mean()
df["range_km"].mean()
df["acceleration_0_100"].mean()
df["expansion_score"].mean()
```

```python
df.groupby("brand")["expansion_score"].mean().sort_values(
    ascending=False
)
```

```python
df.groupby("battery_type")["range_km"].mean()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

plt.scatter(
    df["price_eur"],
    df["range_km"],
    s=df["expansion_score"]*100,
    alpha=0.6
)

for _, row in df.iterrows():
    plt.text(
        row["price_eur"],
        row["range_km"],
        row["brand"]
    )

plt.xlabel("Price (€)")
plt.ylabel("Range (km)")
plt.title("Chinese EV Brands: Price vs Range vs Expansion Score")

plt.show()
```

```python
ranking = df.groupby("brand").agg({
    "range_km":"mean",
    "price_eur":"mean",
    "expansion_score":"mean"
})

ranking
```

```python
ranking.sort_values(
    "expansion_score",
    ascending=False
)
```

```python
df["competitiveness_score"] = (
    df["range_km"]/10
    + df["expansion_score"]*5
    - df["price_eur"]/1000
)

df[[
    "brand",
    "model",
    "competitiveness_score"
]].sort_values(
    "competitiveness_score",
    ascending=False
)
```
